```python
import numpy as np
import tauspec as t
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'cm'
mpl.rcParams['font.size'] = '11'
```

# Calculate xe power spectra

```python
z =  7.5
k=np.logspace(-2, 0)
pxe= t.p_xexe(z, k)
```

# Plot power spectra

```python
plt.plot(k, k**3* pxe, label="z=7.5")


plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"$k$ [1/Mpc]")
plt.ylabel(r"$P_{x_ex_e} \, [(Mpc)^3]$")
plt.xlim(1e-2, 1e0)
plt.legend(loc=0)
```

# Calculate cltau power spectra

```python
l=np.logspace(np.log10(2), np.log10(5000), num=20)
cltau=t.cltau(l)
```

```python
#Plot Cltau
plt.plot(l, l*(l+1)*cltau/2.0/np.pi)
plt.xscale("linear")
plt.yscale("log")
plt.xlabel(r"$\ell$")
plt.ylabel(r"$\ell (\ell +1) C_\ell^{\tau \tau}/2\pi$")
```

```python
# Plot xe 
z=np.linspace(0, 20)
plt.plot(z, t.xe(z))
plt.xscale("linear")
plt.yscale("linear")
plt.xlabel(r"$z$")
plt.ylabel(r"$x_e$")
```
